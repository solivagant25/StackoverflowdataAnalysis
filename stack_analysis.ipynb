```python
import matplotlib.pyplot as plt
import pandas as pd

from collections import Counter
```

```python
data=pd.read_csv('clean.csv')
plt.style.use("fivethirtyeight")
pd.set_option('display.max_columns',85)
fig=plt.figure(figsize=(10,10))
```

```python
ids=data['Respondent']
lang_response=data['LanguageWorkedWith']


lang_counter=Counter()

Languages=[]
People=[]
```

```python
for response in lang_response:
    if isinstance(response, str):
        lang_counter.update(response.split(';'))
for item in lang_counter.most_common(20):
        Languages.append(item[0])
        People.append(item[1])
Languages.reverse()
People.reverse()
    
```

```python

plt.barh(Languages,People,color='green',figure=plt.figure(figsize=(10,10)))
plt.title("Most Popular Languages")
```
